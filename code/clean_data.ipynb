## extract date-time

```python
df['stop_time'] = pd.to_datetime(df['stop_time'],
                                 format="%m/%d/%Y %H:%M",
                                 errors='coerce')
```

```python
print(df['stop_time'].isna().sum().item())
# drop stop_time == na
df = df[~df['stop_time'].isna()]
print(df['stop_time'].isna().sum().item())
```

```python
# seperate columns for date and time
df['date'] = pd.to_datetime(df['stop_time']).dt.date
df['time'] = pd.to_datetime(df['stop_time']).dt.time
```

```python
df['diff'] = pd.to_datetime(df['stop_time']).diff()
df['diff_time'] = df['diff'].apply(
    lambda x: f"{int(x.total_seconds() // 3600):02}:{int((x.total_seconds() % 3600) // 60):02}:{int(x.total_seconds() % 60):02}"
    if pd.notna(x) else None
)
```

```python
df.sort_values(by=['train','direction_id',
       'date', 'stop_time'], inplace=True)
```
